```python
import sys
import os
app_path = os.path.join(os.path.expanduser('~/sandbox/doberman/'))
sys.path.append(app_path)
from doberman import Universe
from doberman import EMA
from doberman import Simulation
from doberman import StockPlot
```

```python
universe = Universe(['jpm'])
universe.load_data()
universe.align_dates('2020-01-01', '2020-12-31')
```

```python
# Calculate EMA signals and plot
ema = EMA(universe.stocks['jpm'])
```

```python
sp = StockPlot(ema.stock_obj)
```
